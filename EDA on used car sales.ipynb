__USED CAR RULE SUMMARY:__  

It requires car dealers to display a window sticker, known as a Buyers Guide, on the used cars they offer for sale. The Buyers Guide discloses whether the dealer offers a warranty and, if so, its terms and conditions, including the duration of the coverage, the percentage of total repair costs the dealer will pay, and which vehicle systems the warranty covers. In states that do not permit sales of used cars “as is,” or without warranties, dealers must display an alternative version of the Buyers Guide.

Most __car dealers__ who sell used vehicles must __comply__ with the __Federal Trade Commission's (FTC's) Used Car Rule__. 

The Used Car Rule applies in all states except Maine and Wisconsin. These two states are exempt because they have similar regulations that require dealers to post disclosures on used vehicles. The Rule applies in the District of Columbia, Puerto Rico, Guam, the U.S. Virgin Islands, and American Samoa.

The booklet defines the Rule's requirements, explains how to prepare and display the Buyers Guide, and offers a compliance checklist.

- Following data processing are being carried out on the data sets:

#### Removing the duplicate rows

````python
# dropping duplicate datasets
data.drop_duplicates (keep = 'first', inplace = True,ignore_index = True)
print('Duplicate rows have been dropped out')
````

__engV__ is given in real number and depicts engine capacity in litres and have specific values. It can't assume intermediate value given by mean.    
Since known car engine capacity in litres less than 0.8 and more than 8.4 are not found, all values outside this range is first imputed with NaN. 

````python
# Checking for data having engV < 0.8 and engV > 8.4
data[(data['engV']<0.8) | (data['engV']>8.4)].groupby(['car','model','engType'])['engV'].agg(['min','median','max'])
````

#### Replacing engV values less than 0.8 & more than 8.4 with NaN

````python
# replacing with NaN weherever engV value is less than 0.8 & greater than 8.4.

data['engV']= data['engV'].apply(lambda x: np.nan if (x>8.4 or x<0.8) else x)
data['engV'].isnull().value_counts()
````

After making __engV__ values less than 0.8 and more than 8.4 as NaN, total missing values become 527 from 434.

Now, using maximum possible matching parameters from __'car'__,__'model'__ and __'engineType'__, __1st mode of 'engV'__ for such a combination is imputed for the missing values. Wherever, no value was available that could match the above parameters, it has been simply imputed with mode of entire __'engV'__ data. 


#### Imputing NaN engV values with the mode of best fit engV in the given Data set

````python
# saving 1st modes for various combinations of 'car','model','engType' to dictionaries

engV_dict  = data.groupby(['car','model','engType'])['engV'].apply(pd.Series.mode).to_dict()
engV_dict1 = data.groupby(['car','model'])['engV'].apply(pd.Series.mode).to_dict()
engV_dict2 = data.groupby(['car'])['engV'].apply(pd.Series.mode).to_dict()
````

````python
engV_null = data[data['engV'].isnull()].index.to_list()  # saving the indexes of null values to a list

c1,c2,c3,c4=0,0,0,0                    #initializing variables for using as counters

#iterating and replacing engV values with 1st mode of corresponding best match for the parameters

for i in engV_null:
    try:        
        data.loc[i,'engV']=engV_dict[(data.loc[i,'car'],data.loc[i,'model'],data.loc[i,'engType'],0)]
        c1+=1
    except:
        try:
            data.loc[i,'engV']=engV_dict1[(data.loc[i,'car'],data.loc[i,'model'],0)]
            c2+=1
        except:
            try:
                data.loc[i,'engV']=engV_dict2[(data.loc[i,'car'],0)]
                c3+=1
            except:
                data.loc[i,'engV']=data['engV'].mode()[0]
                c4+=1                

# printing no. of imputations for each criteria  

print("{}  nos. of engV value imputed with 1st mode matching 'car','model', 'engType' ".format(c1))
print("{}  nos. of engV value imputed with 1st mode matching 'car','model' ".format(c2))
print("{}  nos. of engV value imputed with 1st mode matching 'car' ".format(c3))
print("{}  nos. of engV value imputed with 1st mode matching 'car' ".format(c4))
````

````python
data[data['engV']==0].count()   # checking to confirm 100% imputation
````

- __drive__ is a categorical variable.  
After dropping duplicate rows, there are 510 (5.3%) missing values in __drive__.

Similar to engV, __drive__ using maximum possible matching parameters from 'car','model' and 'engineType', 1st mode of 'drive' for such a combination is imputed for the missing values. Wherever, no value is available that could match the above parameters, it has been simply imputed with mode of entire 'drive' data. 

#### Imputing missing drive values with the mode of best fit drive value in the given Data set

````python
# saving 1st modes for various combinations of 'car','model','engType' to dictionaries

drive_dict = data.groupby(['car','model','engType','engV'])['drive'].apply(pd.Series.mode).to_dict()
drive_dict1 = data.groupby(['car','model','engType'])['drive'].apply(pd.Series.mode).to_dict()
drive_dict2 = data.groupby(['car','model'])['drive'].apply(pd.Series.mode).to_dict()
drive_dict3 = data.groupby(['car'])['drive'].apply(pd.Series.mode).to_dict()
````

````python
drive_null = data[data['drive'].isnull()].index.to_list()  # saving the indexes of null values to a list

c1,c2,c3,c4,c5=0,0,0,0,0                    #initializing variables for using as counters

#iterating and replacing engV values with 1st mode of corresponding best match for the parameters

for i in drive_null:
    try:
        data.loc[i,'drive']=drive_dict[(data.loc[i,'car'],data.loc[i,'model'],data.loc[i,'engType'],data.loc[i,'engV'],0)]
        c1+=1
    except:    
        try:
            data.loc[i,'drive']=drive_dict1[(data.loc[i,'car'],data.loc[i,'model'],data.loc[i,'engType'],0)]
            c2+=1
        except:
            try:
                data.loc[i,'drive']=drive_dict2[(data.loc[i,'car'],data.loc[i,'model'],0)]
                c3+=1
            except:
                try:
                    data.loc[i,'drive']=drive_dict3[(data.loc[i,'car'],0)]
                    c4+=1
                except:
                    data.loc[i,'drive']=data['drive'].mode()[0]
                    c5+=1

# printing no. of imputations for each criteria  

print("{}  nos. of drive value imputed with 1st mode matching 'car','model','engType', 'engV' ".format(c1))
print("{}  nos. of drive value imputed with 1st mode matching 'car','model','engType' ".format(c2))
print("{}  nos. of drive value imputed with 1st mode matching 'car','model' ".format(c3))
print("{}  nos. of drive value imputed with 1st mode matching 'car' ".format(c4))
print("{}  nos. of drive value imputed with 1st mode of overall data ".format(c5))
print("{} Total nos. of drive value imputed wherever null  ".format(c1+c2+c3+c4+c5))
````

#### Imputing mileage for the year 2016

````python
# replacing all zeroes in mileage for year 2016 with mean for non zero values
data.loc[data['year']==2016,'mileage'] = data.loc[data['year']==2016,'mileage'].apply(lambda x: 4.27 if x==0 else x)
````

````python
data[data['year']==2016]['mileage'].agg(['count','mean','median','min','max','sum']) # checking mileage stats for 2016 after imputation
````

#### Replacing zero values in mileage with median of mileage for the years other than  2016

````python
year_list.remove(2016)    # removing 2016 from the list as its zeroes have been replaced by its mean

#iterating over the list of years for which mileage had zeroes and replacing zeroes with the respective median for the year

for i in year_list:    
    data.loc[data['year']==i,'mileage'] = data.loc[data['year']==i,'mileage'].replace(0,data.loc[data['year']==i,'mileage'].median())
````

#### Imputing zero price values with the median of best fit price value in the given Data set

````python
# saving medians of price for various combinations of 'car','model','engType', 'drive', 'year' to dictionaries

price_dict  = data.groupby(['car','year','model','engV','engType','drive'])['price'].median().to_dict()
price_dict1 = data.groupby(['car','year','model','engV','engType'])['price'].median().to_dict()
price_dict2 = data.groupby(['car','year','model','engV'])['price'].median().to_dict()
price_dict3 = data.groupby(['car','year','model'])['price'].median().to_dict()
price_dict4 = data.groupby(['car','year'])['price'].median().to_dict()
price_dict5 = data.groupby(['car'])['price'].median().to_dict()

# removing zero value items from the dictionary to ensure that there are no medians with zero value
price_dict  = {x:y for x,y in price_dict.items()  if y!=0}
price_dict1 = {x:y for x,y in price_dict1.items() if y!=0}
price_dict2 = {x:y for x,y in price_dict2.items() if y!=0}
price_dict3 = {x:y for x,y in price_dict3.items() if y!=0}
price_dict4 = {x:y for x,y in price_dict4.items() if y!=0}
price_dict5 = {x:y for x,y in price_dict5.items() if y!=0}

price_zero = data[data['price']==0.0].index.to_list()  # saving the indexes of zero values to a list
print("Total zeroes in price = ",len(price_zero)) 
````

````python

c1,c2,c3,c4,c5,c6,c7=0,0,0,0,0,0,0                    #initializing variables for using as counters

#iterating and replacing engV values with 1st mode of corresponding best match for the parameters

for i in price_zero:
    try:
        data.loc[i,'price']=price_dict[(data.loc[i,'car'],data.loc[i,'year'],data.loc[i,'model'],data.loc[i,'engV'],data.loc[i,'engType'],data.loc[i,'drive'])]
        c1+=1
    except:
        try:
            data.loc[i,'price']=price_dict1[(data.loc[i,'car'],data.loc[i,'year'],data.loc[i,'model'],data.loc[i,'engV'],data.loc[i,'engType'])]
            c2+=1
        except:
            try:
                data.loc[i,'price']=price_dict2[(data.loc[i,'car'],data.loc[i,'year'],data.loc[i,'model'],data.loc[i,'engV'])]
                c3+=1
            except:
                try:
                    data.loc[i,'price']=price_dict3[(data.loc[i,'car'],data.loc[i,'year'],data.loc[i,'model'])]
                    c4+=1
                except:                
                    try:
                        data.loc[i,'price']=price_dict4[(data.loc[i,'car'],data.loc[i,'year'])]
                        c5+=1
                    except: 
                        try:
                            data.loc[i,'price']=price_dict5[(data.loc[i,'car'])] 
                            c6+=1
                        except:
                            data.loc[i,'price']=data['price'].median()
                            c7+=1                   
                        

# printing no. of imputations for each criteria  

print("{}  nos. of price value imputed with median of price for combination of 'car','year','model',engV',engType','drive' ".format(c1))
print("{}  nos. of price value imputed with median of price for combination of 'car','year','model',engV',engType' ".format(c2))
print("{}  nos. of price value imputed with median of price for combination of 'car','year','model',engV' ".format(c3))
print("{}  nos. of price value imputed with median of price for combination of 'car','year','model' ".format(c4))
print("{}  nos. of price value imputed with median of price for combination of 'car','year', ".format(c5))
print("{}  nos. of price value imputed with median of price for combination of 'car' ".format(c6))
print("{}  nos. of price value imputed with median of 'price' ".format(c7))
````

- Cars are categorised based on 95th percentile value. Car make having 95th percentile of its price above 2 standard deviation of the overall price is categorised as Luxury Car and remaining as Economy Car 

````python
std_2 = (data['price'].std())*2           # Twice of standard deviation of price

print('2 Standard deviation of price is   ',std_2,'\n')

# creating dictionaty of 95th percentile of price for each car make
car_dict=data.groupby('car')['price'].quantile(q=0.95).to_dict()  

# categorising car as luxury if 95% quartile of price is more than 2 std deviations of overall price & saving to list
luxury_car = list({x:y for x,y in car_dict.items() if y>std_2})
   
print('List of Luxury cars is given below : \n \n', luxury_car)
print(' \n No. of Luxury car make are :  ', len(luxury_car))
````

### Dealing with categorical variables
---


- Used car dataset has following categorical variables:  
    - car
    - model
    - body
    - engType
    - drive
    - registration
    
We know that machine understands only numbers, unlike human brain it is inefficient to categorize strings.
The solution that tackles this problem is encoding the categorical data,
"Encoding" helps machine understand the meaning behind the features.

One such encoding technique is "Label Encoding".

"Label Encoding" works on a very simple approach of assigning integers to categories in the features.
Integers are assigned on the basis of acending order of the characters in the categories.
for example:
if we have three categories namely red, blue and green. Then after encoding them with LabelEncoder we get
red as 2, blue as 0 and green as 1.
